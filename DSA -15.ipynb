Given an arrayÂ **arr[ ]**Â of sizeÂ **N**Â havingÂ elements, theÂ task is to find the next greater element for each element of the array in order of their appearance in the array.Next greater element of an element in the array is the nearest element on the right which is greater than the current element.If there does not exist next greater of current element, then next greater element for current element is -1. For example, next greater of the last element is always -1.

**Example 1:**

```
Input:
N = 4, arr[] = [1 3 2 4]
Output:
3 4 4 -1
Explanation:
In the array, the next larger element
to 1 is 3 , 3 is 4 , 2 is 4 and for 4 ?
since it doesn't exist, it is -1.

```

**Example 2:**

```
Input:
N = 5, arr[] [6 8 0 1 3]
Output:
8 -1 1 3 -1
Explanation:
In the array, the next larger element to
6 is 8, for 8 there is no larger elements
hence it is -1, for 0 it is 1 , for 1 it
is 3 and then for 3 there is no larger
element on right and hence -1.
```

</aside>

**Solution 1:**

To solve this problem, we can use a stack. We iterate through the array from right to left and keep track of the elements in the stack. For each element, we pop elements from the stack until we find an element greater than the current element or the stack becomes empty. The element at the top of the stack after this process is the next greater element for the current element.

Here's the implementation in Python:

```python
def next_greater_element(arr):
    stack = []
    result = []
    
    # Iterate through the array from right to left
    for i in range(len(arr)-1, -1, -1):
        # Pop elements from the stack until we find a greater element or the stack becomes empty
        while stack and stack[-1] <= arr[i]:
            stack.pop()
        
        # Append the next greater element to the result
        if stack:
            result.append(stack[-1])
        else:
            result.append(-1)
        
        # Push the current element to the stack
        stack.append(arr[i])
    
    # Reverse the result to get the order of appearance
    result.reverse()
    
    return result

# Test the function
arr = [1, 3, 2, 4]
print(next_greater_element(arr))  # Output: [3, 4, 4, -1]

arr = [6, 8, 0, 1, 3]
print(next_greater_element(arr))  # Output: [8, -1, 1, 3, -1]
```

**Solution 2:**

To find the nearest smaller number on the left side for each element, we can use a stack. We iterate through the array from left to right and keep track of the elements in the stack. For each element, we pop elements from the stack until we find a smaller element or the stack becomes empty. The element at the top of the stack after this process is the nearest smaller number for the current element. If the stack becomes empty, it means there is no smaller number on the left.

Here's the implementation in Python:

```python
def nearest_smaller_number(arr):
    stack = []
    result = []
    
    # Iterate through the array from left to right
    for i in range(len(arr)):
        # Pop elements from the stack until we find a smaller element or the stack becomes empty
        while stack and stack[-1] >= arr[i]:
            stack.pop()
        
        # Append the nearest smaller number to the result
        if stack:
            result.append(stack[-1])
        else:
            result.append(-1)
        
        # Push the current element to the stack
        stack.append(arr[i])
    
    return result

# Test the function
arr = [1, 6, 2]
print(nearest_smaller_number(arr))  # Output: [-1, 1, 1]

arr = [1, 5, 0, 3, 4, 5]
print(nearest_smaller_number(arr))  # Output: [-1, 1, -1, 0, 3, 4]
```

```
Input:
push(2)
pop()
pop()
push(3)
Output:2 -1
```

**Solution 3:**

To implement a stack using two queues, we can use the following approach:

- We maintain two queues, `q1` and `q2`.
- The `push` operation is performed by enqueueing the element into `q1`.
- The `pop` operation is performed by dequeueing all elements from `q1` except the last one, and enqueueing them into `q2`. The last element in `q1` is then dequeued, which simulates the behavior of a stack.
- To switch the roles of `q1` and `q2` for the next operation, we can swap the references of `q1` and `q2`.

Here's the implementation in Python:

```python
from queue import Queue

class Stack:
    def __init__(self):
        self.q1 = Queue()
        self.q2 = Queue()
    
    def push(self, val):
        self.q1.put(val)
    
    def pop(self):
        if self.q1.empty():
            return -1
        
        while self.q1.qsize() > 1:
            self.q2.put(self.q1.get())
        
        val = self.q1.get()
        
        self.q1, self.q2 = self.q2, self.q1
        
        return val

# Test the implementation
stack = Stack()
stack.push(2)
stack.push(3)
print(stack.pop())  # Output: 3
stack.push(4)
print(stack.pop())  # Output: 4
```

**Solution 5:**

To reverse a string using a stack, we can iterate through the characters of the string and push them onto the stack. Then, we pop the characters from the stack and concatenate them to form the reversed string.

Here's the implementation in Python:

```python
def reverse_string(s):
    stack = []
    
    # Push characters onto the stack
    for char in s:
        stack.append(char)
    
    reversed_string = ""
    
    # Pop characters from the stack and concatenate them
    while stack:
        reversed_string += stack.pop()
    
    return reversed_string

# Test the function
s = "GeeksforGeeks"
print(reverse_string(s))  # Output: skeeGrofskeeG
```

Given stringÂ **S**Â representing aÂ postfix expression, the task is to evaluate the expression and find the final value. Operators will only include the basic arithmetic operators likeÂ ***, /, + and -**.

**Example 1:**

```
Input: S = "231*+9-"
Output: -4
Explanation:
After solving the given expression,
we have -4 as result.

```

**Example 2:**

```
Input: S = "123+*8-"
Output: -3
Explanation:
After solving the given postfix
expression, we have -3 as result.
```

**Solution 6:**

To evaluate a postfix expression, we can use a stack. We iterate through the characters of the expression and perform the following steps:

- If the character is a digit, we convert it to an integer and push it onto the stack.
- If the character is an operator, we pop two operands from the stack, perform the operation, and push the result back onto the stack.

After iterating through all the characters, the final result will be the top element of the stack.

Here's the implementation in Python:

```python
def evaluate_postfix(expression):
    stack = []
    
    for char in expression:
        if char.isdigit():
            stack.append(int(char))
        else:
            operand2 = stack.pop()
            operand1 = stack.pop()
            
            if char == '+':
                result = operand1 + operand2
            elif char == '-':
                result = operand1 - operand2
            elif char == '*':
                result = operand1 * operand2
            elif char == '/':
                result = operand1 / operand2
            
            stack.append(result)
    
    return stack.pop()

# Test the function
expression = "231*+9-"
print(evaluate_postfix(expression))  # Output: -4

expression = "123+*8-"
print(evaluate_postfix(expression))  # Output: -3
```

Implement theÂ `MinStack`Â class:

- `int getMin()`Â retrieves the minimum element in the stack.

You must implement a solution withÂ `O(1)`Â time complexity for each function.

**Example 1:**

```
Input
["MinStack","push","push","push","getMin","pop","top","getMin"]
[[],[-2],[0],[-3],[],[],[],[]]

Output
[null,null,null,null,-3,null,0,-2]

Explanation
MinStack minStack = new MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.getMin(); // return -3
minStack.pop();
minStack.top();    // return 0
minStack.getMin(); // return -2
```

**Solution 7:**

To implement a stack that supports push, pop, top, and retrieving the minimum element in constant time, we can use an additional stack to keep track of the minimum elements.

Here's the implementation in Python:

```python
class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []
    
    def push(self, val):
        self.stack.append(val)
        
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)
    
    def pop(self):
        if self.stack:
            val = self.stack.pop()
            
            if val == self.min_stack[-1]:
                self.min_stack.pop()
    
    def top(self):
        if self.stack:
            return self.stack[-1]
    
    def getMin(self):
        if self.min_stack:
            return self.min_stack[-1]

# Test the implementation
minStack = MinStack()
minStack.push(-2)
minStack.push(0)
minStack.push(-3)
print(minStack.getMin())  # Output: -3
minStack.pop()
print(minStack.top())     # Output: 0
print(minStack.getMin())  # Output: -2
```

**Solution 8:**

To compute the amount of water that can be trapped after raining, we can use the two-pointer approach. We initialize two pointers, `left` and `right`, at the leftmost and rightmost bars of the elevation map. We also maintain two variables, `max_left` and `max_right`, to keep track of the maximum height encountered from the left and right sides.

We iterate through the elevation map from both directions, moving the pointers inward. At each step, we compare the heights of the bars at the `left` and `right` pointers. If the height at the `left` pointer is less than or equal to the height at the `right` pointer, we update the `max_left` variable and calculate the trapped water on the left side. Otherwise, we update the `max_right` variable and calculate the trapped water on the right side.

Here's the implementation in Python:

```python
def trap_water(height):
    n = len(height)
    left = 0
    right = n - 1
    max_left = 0
    max_right = 0
    water = 0
    
    while left < right:
        if height[left] <= height[right]:
            if height[left] >= max_left:
                max_left = height[left]
            else:
                water += max_left - height[left]
            left += 1
        else:
            if height[right] >= max_right:
                max_right = height[right]
            else:
                water += max_right - height[right]
            right -= 1
    
    return water

# Test the function
height = [0,1,0,2,1,0,1,3,2,1,2,1]
print(trap_water(height))  # Output: 6
```

This solution has a time complexity of O(n) and a space complexity of O(1).

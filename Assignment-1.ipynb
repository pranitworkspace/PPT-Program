1. Write a Python program to reverse a string without using any built-in string reversal functions.
```python
def reverse_string(s):
    reversed_str = ""
    for char in s:
        reversed_str = char + reversed_str
    return reversed_str

# Example usage
string = "Hello, World!"
reversed_string = reverse_string(string)
print(reversed_string)  # Output: "!dlroW ,olleH"
```

4. Implement a function to count the occurrence of each element in a list.
```python
from collections import defaultdict

def count_occurrences(lst):
    occurrence_count = defaultdict(int)
    for item in lst:
        occurrence_count[item] += 1
    return occurrence_count

# Example usage
numbers = [1, 2, 3, 2, 1, 3, 4, 1]
occurrence_count = count_occurrences(numbers)
print(occurrence_count)  # Output: defaultdict(<class 'int'>, {1: 3, 2: 2, 3: 2, 4: 1})
```

5. Write a Python program to find the second largest number in a list.
```python
def find_second_largest(lst):
    largest = max(lst)
    second_largest = float('-inf')
    for num in lst:
        if num != largest and num > second_largest:
            second_largest = num
    return second_largest

# Example usage
numbers = [5, 9, 3, 7, 2]
second_largest = find_second_largest(numbers)
print(second_largest)  # Output: 7
```

15. Write a program to find the median of a list of numbers.
```python
def find_median(lst):
    sorted_lst = sorted(lst)
    n = len(sorted_lst)
    if n % 2 == 0:
        median = (sorted_lst[n // 2 - 1] + sorted_lst[n // 2]) / 2
    else:
        median = sorted_lst[n // 2]
    return median

# Example usage
numbers = [4, 5, 2, 8, 3, 9, 1, 6, 7]
median = find_median(numbers)
print(median)  # Output: 5
```

18. Implement a function to find the maximum subarray sum in a given list.
```python
def max_subarray_sum(lst):
    max_sum = float('-inf')
    current_sum = 0
    for num in lst:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)
    return max_sum

# Example usage
numbers = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
subarray_sum = max_subarray_sum(numbers)
print(subarray_sum)  # Output: 6
```

22. Implement a function to find the first non-repeating character in a string.
```python
def find_first_non_repeating_character(s):
    char_count = {}
    for char in s:
        char_count[char] = char_count.get(char, 0) + 1
    for char in s:
        if char_count[char] == 1:
            return char
    return None

# Example usage
string = "abracadabra"
first_non_repeating_char = find_first_non_repeating_character(string)
print(first_non_repeating_char)  # Output: "c"
```

23. Write a program to find the prime factors of a given number.
```python
def prime_factors(n):
    factors = []
    divisor = 2
    while divisor <= n:
        if n % divisor == 0:
            factors.append(divisor)
            n /= divisor
        else:
            divisor += 1
    return factors

# Example usage
number = 84
factors = prime_factors(number)
print(factors)  # Output: [2, 2, 3, 7]
```

25. Write a Python program to merge two sorted lists into a single sorted list.
```python
def merge_sorted_lists(lst1, lst2):
    merged_list = []
    i, j = 0, 0
    while i < len(lst1) and j < len(lst2):
        if lst1[i] < lst2[j]:
            merged_list.append(lst1[i])
            i += 1
        else:
            merged_list.append(lst2[j])
            j += 1


    merged_list.extend(lst1[i:])
    merged_list.extend(lst2[j:])
    return merged_list

# Example usage
list1 = [1, 3, 5]
list2 = [2, 4, 6]
merged_list = merge_sorted_lists(list1, list2)
print(merged_list)  # Output: [1, 2, 3, 4, 5, 6]
```

26. Implement a function to find the mode of a list of numbers.
```python
from collections import Counter

def find_mode(lst):
    counter = Counter(lst)
    max_count = max(counter.values())
    modes = [num for num, count in counter.items() if count == max_count]
    return modes

# Example usage
numbers = [1, 2, 3, 2, 1, 3, 4, 1]
mode = find_mode(numbers)
print(mode)  # Output: [1]
```

28. Implement a function to calculate the square root of a given number.
```python
def square_root(n):
    guess = n
    while True:
        next_guess = (guess + n / guess) / 2
        if abs(next_guess - guess) < 1e-10:
            return next_guess
        guess = next_guess

# Example usage
number = 16
square_root_result = square_root(number)
print(square_root_result)  # Output: 4.0
```

30. Implement a function to find the minimum element in a rotated sorted list.
```python
def find_minimum(lst):
    left = 0
    right = len(lst) - 1
    while left < right:
        mid = (left + right) // 2
        if lst[mid] > lst[right]:
            left = mid + 1
        else:
            right = mid
    return lst[left]

# Example usage
numbers = [4, 5, 6, 7, 0, 1, 2]
minimum = find_minimum(numbers)
print(minimum)  # Output: 0
```

```python
def power(base, exponent):
    if exponent == 0:
        return 1
    elif exponent < 0:
        return 1 / power(base, -exponent)
    else:
        return base * power(base, exponent - 1)

# Example usage
base = 2
exponent = 3
result = power(base, exponent)
print(result)  # Output: 8
```

34. Implement a function to find the longest common prefix among a list of strings.
```python
def find_longest_common_prefix(strs):
    if not strs:
        return ""
    shortest = min(strs, key=len)
    for i, char in enumerate(shortest):
        for other in strs:
            if other[i] != char:
                return shortest[:i]
    return shortest

# Example usage
strings = ["flower", "flow", "flight"]
longest_common_prefix = find_longest_common_prefix(strings)
print(longest_common_prefix)  # Output: "fl"
```

41. Write a Python program to find the smallest missing positive integer in a list.
```python
def find_smallest_missing_positive(lst):
    n = len(lst)
    for i in range(n):
        while 1 <= lst[i] <= n and lst[i] != lst[lst[i] - 1]:
            lst[lst[i] - 1], lst[i] = lst[i], lst[lst[i] - 1]
    for i in range(n):
        if lst[i] != i + 1:
            return i + 1
    return n + 1

# Example usage
numbers = [3, 4, -1, 1]
smallest_missing_positive = find_smallest_missing_positive(numbers)
print(smallest_missing_positive)  # Output: 2
```

42. Implement a function to find the longest palindrome substring in a given string.
```python
def longest_palindrome(s):
    longest = ""
    for i in range(len(s)):
        for j in range(i, len(s)):
            substring = s[i:j + 1]
            if substring == substring[::-1] and len(substring) > len(longest):
                longest = substring
    return longest

# Example usage
string = "babad"
longest_palindrome_substring = longest_palindrome(string)
print(longest_palindrome_substring)  # Output: "bab"
```

Given an integer¬†`n`, return¬†*`true`¬†if it is a power of two. Otherwise, return¬†`false`*.

An integer¬†`n`¬†is a power of two, if there exists an integer¬†`x`¬†such that¬†`n == 2x`.

**Example 1:**
Input: n = 1 

Output: true

**Example 2:**
Input: n = 16 

Output: true

**Example 3:**
Input: n = 3 

Output: false

</aside>

**Answer 1:**

We can solve this problem by checking if the given number `n` is greater than 0 and at the same time, its bitwise AND operation with `(n - 1)` is equal to 0. This condition checks if `n` has only one bit set, which is the characteristic of a power of two.

Here's the implementation in Python:

```python
def isPowerOfTwo(n):
    if n <= 0:
        return False
    return (n & (n - 1)) == 0
```

üí° **Question 3**

****Given a positive integer, N. Find the factorial of N. 

**Example 1:**

Input: N = 5 

Output: 120

**Example 2:**

Input: N = 4

Output: 24

</aside>

**Answer 3:**

We can calculate the factorial of a positive integer `N` by using a recursive function. The factorial of `N` is the product of all positive integers from `1` to `N`.

Here's the recursive implementation in Python:

```python
def factorial(N):
    if N == 0 or N == 1:
        return 1
    return N * factorial(N - 1)
```

üí° **Question 5**

Given an array of integers¬†**arr**, the task is to find maximum element of that array using recursion.

**Example 1:**

Input: arr = {1, 4, 3, -5, -4, 8, 6};
Output: 8

**Example 2:**

Input: arr = {1, 4, 45, 6, 10, -8};
Output: 45

</aside>

**Answer 5:**

We can find the maximum element of an array using recursion by comparing each element with the maximum of the remaining subarray.

Here's the recursive implementation in Python:

```python
def findMax(arr):
    if len(arr) == 1:
        return arr[0]
    else:
        return max(arr[0], findMax(arr[1:]))
```

üí° **Question 6**

Given first term (a), common difference (d) and a integer N of the Arithmetic Progression series, the task is to find Nth term of the series.

**Example 1:**

Input : a = 2 d = 1 N = 5
Output : 6
The 5th term of the series is : 6

**Example 2:**

Input : a = 5 d = 2 N = 10
Output : 23
The 10th term of the series is : 23

</aside>

**Answer 6:**

The Nth term of an Arithmetic Progression (AP) series can be calculated using the formula: `NthTerm = a + (N - 1) * d`, where `a` is the first term and `d` is the common difference.

Here's the implementation in Python:

```python
def findNthTerm(a, d, N):
    return a + (N - 1) * d
```

üí° **Question 7**

Given a string S, the task is to write a program to print all permutations of a given string.

**Example 1:**

***Input:***

*S = ‚ÄúABC‚Äù*

***Output:***

*‚ÄúABC‚Äù, ‚ÄúACB‚Äù, ‚ÄúBAC‚Äù, ‚ÄúBCA‚Äù, ‚ÄúCBA‚Äù, ‚ÄúCAB‚Äù*

**Example 2:**

***Input:***

*S = ‚ÄúXY‚Äù*

***Output:***

*‚ÄúXY‚Äù, ‚ÄúYX‚Äù*

</aside>

**Answer 7:**

We can use the `itertools` module in Python to generate all permutations of a given string.

Here's the implementation:

```python
import itertools

def printPermutations(S):
    permutations = list(itertools.permutations(S))
    for permutation in permutations:
        print(''.join(permutation))
```

üí° **Question 8**

Given an array, find a product of all array elements.

**Example 1:**

Input  : arr[] = {1, 2, 3, 4, 5}
Output : 120
**Example 2:**

Input  : arr[] = {1, 6, 3}
Output : 18

</aside>

**Answer 8:**

To find the product of all elements in an array, we can iterate over the array and multiply each element.

Here's the implementation in Python:

```python
def productOfArray(arr):
    product = 1
    for num in arr:
        product *= num
    return product
```
